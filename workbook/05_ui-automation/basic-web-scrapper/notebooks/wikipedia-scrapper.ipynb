```python
import requests
import bs4
```

```python
webpage = requests.get("https://en.wikipedia.org/wiki/Grace_Hopper")
```

```python
soup = bs4.BeautifulSoup(webpage.text, "lxml")
```

```python
main_heading = soup.select("h1")
print(main_heading[0].getText())
```

```python
paragraphs = soup.select("p")
for paragraph in paragraphs:
    print(paragraph)
```

```python
table_of_content_items = soup.select(".vector-toc-text .vector-toc-numb + span")
for table_of_content_item in table_of_content_items:
    print(table_of_content_item.getText())
```

```python
thumbnail_image = soup.select(".infobox-image .mw-file-element")
image_src = thumbnail_image[0].get("src")
print(f"https:{image_src}")
```

```python

image_link = requests.get(f"https:{image_src}")
with open("grace_hopper.jpg", "wb") as file:
    file.write(image_link.content)
```
