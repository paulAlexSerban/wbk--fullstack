```python
from PIL import Image
```

```python
words = Image.open('./images/word_matrix.png')
mask = Image.open('./images/mask.png')
```

```python
#resize to match
mask_size = mask.size
words_size = words.size

mask = mask.resize(words_size)
```

```python
# make mask transparent
mask = mask.convert("RGBA")
mask.putalpha(100)  # Set transparency level (0-255)
mask
```

```python
# overlay the images
words.paste(mask, (0, 0), mask)
```
