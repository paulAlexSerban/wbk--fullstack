```python
import requests
import bs4
```

```python
def fetch_page_content(url: str) -> str:
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        return ""
```

```python
def fetch_paginated_content(base_url: str, page_number: int) -> str:
    url = f"{base_url}/page/{page_number}"
    return fetch_page_content(url)
```

```python
# fetch each page until response is empty string
page_number = 1
authors_set = set()
tags_set = set()
while True:
    content = fetch_paginated_content("https://quotes.toscrape.com", page_number)
    soup = bs4.BeautifulSoup(content, "lxml")
    authors = [author.text for author in soup.select(".author")]
    tags = [tag.text.strip() for tag in soup.select(".tag-item")]
    if len(authors) == 0:
        break
    authors_set.update(authors)
    tags_set.update(tags)
    page_number += 1
    
print("Authors:")
for author in authors_set:
    print(author)

print("\nTags:")
for tag in tags_set:
    print(tag)
```
