```python
import openpyxl
from openpyxl import Workbook
import random
import lorem
```

```python
EXCEL_FILE_NAME = 'output.xlsx'
WORKSHEET_TITLE = "seeded_data"
WORKSHEET_HEADER = ['slug', 'label', 'feature-key']

FRUITS = ['Apple', 'Banana', 'Cherry', 'Date', 'Elderberry', 'Fig', 'Grape', 'Honeydew']
COLOR_FEATURES = ['red', 'green', 'blue', 'yellow', 'purple', 'orange', 'pink', 'brown']

ROW_COUNT = 200
```

```python
def get_random_lorem_slug():
    words = lorem.sentence().split()[:4]
    return '-'.join(words).lower()
```

```python
wb = Workbook()
ws = wb.active

ws.title = WORKSHEET_TITLE
ws.append(WORKSHEET_HEADER)

for _ in range(ROW_COUNT):
    slug = get_random_lorem_slug()
    fruit = random.choice(FRUITS)
    colors = ', '.join([f"[{i}, feature-{color}]" for i, color in enumerate(random.sample(COLOR_FEATURES, 3))])
    ws.append([slug, fruit, colors])

wb.save(EXCEL_FILE_NAME)
print(f"Excel file '{EXCEL_FILE_NAME}' created successfully.")
```
