```python
import pypdf
import re
file = open('./Find_the_Phone_Number.pdf', 'rb')  # rb stands for "read binary" it is necessary for PDF files

pdf = pypdf.PdfReader(file)
page_count = len(pdf.pages)

phone_numbers = []
pattern = r'\d{3}.\d{3}.\d{4}'  # Regular expression to match phone numbers in the format xxx.xxx.xxxx
for page in pdf.pages:
    text = page.extract_text()
    numbers = re.findall(pattern, text)
    phone_numbers.extend(numbers)


file.close()
print(phone_numbers)
```
