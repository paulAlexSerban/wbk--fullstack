```python
import pypdf
```

```python
file = open('./Some_BrandNew_Doc.pdf', 'rb') # rb stands for "read binary" it is necessary for PDF files
pdf_reader = pypdf.PdfReader(file)
page_one = pdf_reader.pages[0]
print(page_one.extract_text())
```

```python
file.close()
```

```python
second_file = open('./Some_BrandNew_Doc.pdf', 'rb')
pdf_reader = pypdf.PdfReader(second_file)
first_page = pdf_reader.pages[0]
```

```python
pdf_writer = pypdf.PdfWriter()
```

```python
pdf_writer.add_page(first_page)
```

```python
pdf_output = open('./output.pdf', 'wb') # wb stands for "write binary"
pdf_writer.write(pdf_output)

pdf_output.close()
second_file.close()
file.close()
```
