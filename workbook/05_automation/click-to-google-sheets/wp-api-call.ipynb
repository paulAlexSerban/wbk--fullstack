```python
# ttbrowsert0ken

import json
from datetime import datetime
import requests

def log_event(src, client_ts):
    url = "https://3sof.com/wp-json/tracker/v1/log"
    headers = {"Content-Type": "application/json"}
    data = {
        "src": src,
        "client_ts": client_ts
    }
    
    response = requests.post(url, headers=headers, data=json.dumps(data))
    
    if response.status_code == 200:
        print("Event logged successfully.")
    else:
        print(f"Failed to log event. Status code: {response.status_code}, Response: {response.text}")

if __name__ == "__main__":
    src = "test-local-script"
    client_ts = datetime.utcnow().isoformat() + "Z"  # Current UTC time in ISO format
    log_event(src, client_ts)
```
