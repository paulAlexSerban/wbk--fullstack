```python
import requests
import json
import datetime

def post_to_sheets():
    url = "https://script.google.com/macros/s/AKfycbxK9U_7UwA9eX3ui5jDdl8gs7VkGBvWPVKOmPm7uMwjGx69bNkLSBWX_DZ2-9COnkKj/exec"

    data = {
        "client_ts": datetime.datetime.now().isoformat(),
        "src": "example_src",
        "utm_source": "example_utm_source",
        "utm_medium": "example_utm_medium",
        "utm_campaign": "example_utm_campaign",
        "utm_content": "example_utm_content",
        "utm_term": "example_utm_term",
        "cid": "example_cid",
        "eid": "example_eid",
        "lid": "example_lid",
        "destination": "example_destination",
        "request_id": "example_request_id",
        "referrer": "example_referrer",
        "user_agent": "example_user_agent",
        "language": "en-US",
        "timezone": "UTC",
        "key": "t3st@track3e@2026"
    }


    headers = {
        "Content-Type": "application/json"
    }

    response = requests.post(url, headers=headers, data=json.dumps(data))
    if response.status_code == 200:
        print({"status": "success", "message": "Data posted successfully!", "response": response.text})
        print("Data posted successfully!")
    else:
        print(f"Failed to post data. Status code: {response.status_code}")

if __name__ == "__main__":    
    post_to_sheets()
```
