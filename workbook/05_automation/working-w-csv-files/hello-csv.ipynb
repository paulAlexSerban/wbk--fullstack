```python
import csv
```

```python
data = open('./example.csv', encoding='utf-8')
```

```python
csv_data = csv.reader(data)
```

```python
data_lines = list(csv_data)
```

```python
column_names = data_lines[0]
data_rows = data_lines[1:]
```

```python
all_emails = []
full_names = []
for row in data_rows:
    all_emails.append(row[3])
    full_names.append(row[1] + ' ' + row[2])
    
print(all_emails)
print(full_names)
```

```python
file_to_output = open('./output.csv', mode='w', newline='', encoding='utf-8')
csv_writer = csv.writer(file_to_output)
csv_writer.writerow(['Full Name', 'Email'])
for i in range(len(full_names)):
    csv_writer.writerow([full_names[i], all_emails[i]])
file_to_output.close() 
```
